```
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go 
from io import StringIO
from os import path
```

```
locs = '''Lat,Lon,Name
48.000000, -97.000000, pt1
48.000000, -96.000000, pt2
47.000000, -95.000000, pt3'''
```

```

windowspath = r'C:\Users\David\Google Drive\python\put_shape_on_map/driving_radii.csv'
linuxpath  = '/mnt/chromeos/GoogleDrive/MyDrive/python/put_shape_on_map/driving_radii.csv'
if path.exists(linuxpath):
    csvdata = linuxpath
elif path.exists(windowspath):
    csvdata = windowspath
else:
    csvdata = StringIO(locs)
```

```
df = pd.read_csv(csvdata)
latc = df.Lat.mean()
lonc = df.Lon.mean()
hours = df.Hours.unique()
```

```

def scatter_mapbox_hour(df, hour):
    dff = df.where(df.Hours == hour).dropna(thresh=1)
    name = '1 Hour' if hour == 1 else '{} Hours'.format(hour)
    return go.Scattermapbox(lon=dff.Lon, lat=dff.Lat, fill="toself", name=name)

gos = [scatter_mapbox_hour(df,hour) for hour in df.Hours.unique()]
len(gos)
```

```
fig = go.Figure(gos)
fig.update_layout(
    autosize=True,
    hovermode='closest',
    mapbox=dict(
        style="open-street-map",
        zoom=4,
        center = go.layout.mapbox.Center(
            lat=latc,
            lon=lonc
        )
    )
)
fig.write_html('map.html')
fig.show()
```
